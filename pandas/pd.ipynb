```python
#zip 
item1 = ['macbook air m4', 'nitro v15', 'victus']
item2 = ['apple', 'acer', 'hp']
item3 = [50000, 30000, 25000]

cols = ['name', 'brand', 'price']
data = list(zip(item1,item2,item3))
df = pd.DataFrame(data, columns=cols)
df
```

```python
de = pd.read_excel("Employee.xlsx", index_col="Name")
de
```

```python
de.isnull().any()
```

```python
de.isnull().sum()
```

```python
de.describe()
```

```python
#แทนค่าข้อมูลที่สูญหายด้วยค่าเฉลี่ย
select = "Salary"
de[select] = de[select].fillna(de.describe().mean())
```

```python
de
```

```python
#แทนค่าข้อมูลที่สูญหายด้วยค่ากำหนดเอง
select = "Salary"
de[select] = de[select].fillna(18000)
```

```python
de
```

```python
#แทนค่าข้อมูลด้วยค่าก่อนหน้า
de.fillna(method='pad', inplace=True)
```

```python
de
```

```python
select = 'Address'
de[select] = de[select].fillna('intern')
```

```python
de
```

```python
#แทนที่ค่าว่างด้วยค่าถัดไป
de.fillna(method='bfill', inplace=True)
```

```python
de
```

```python
#ลบแถวทั้งหมดที่มีค่าว่าง
de.dropna(inplace=True)
```

```python
de
```

```python
#ลบแถวบางส่วนที่มีค่าว่าง
de.dropna(subset=['Age','Job','Salary'], inplace=True)
de
```

```python
#ลบคอลัมน์ที่จ้อมูลไม่รคบ
de.dropna(axis='columns', inplace=True)
de
```
