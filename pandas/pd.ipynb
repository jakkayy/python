```python
data = [
    ['macbook air m4', 'apple', 50000],
    ['nitro v15', 'acer', 30000],
    ['victus', 'hp', 25000]
]

cols = ['name', 'brand', 'price']
df = pd.DataFrame(data, columns=cols)
df
```

```python
#zip 
item1 = ['macbook air m4', 'nitro v15', 'victus']
item2 = ['apple', 'acer', 'hp']
item3 = [50000, 30000, 25000]

cols = ['name', 'brand', 'price']
data = list(zip(item1,item2,item3))
df = pd.DataFrame(data, columns=cols)
df
```

```python
#dataframe_from_dic
data = [
    {'Name' : 'macbook', 'brand' : 'apple', 'price' : 40000},
    {'Name' : 'nitro', 'brand' : 'acer', 'price' : 30000},
    {'Name' : 'victus', 'brand' : 'hp', 'price' : 25000}
]

df = pd.DataFrame(data)
#เปลี่ยน index ใหม่แบบถาวร
df.set_index(['Name'], inplace=True)
df
```

```python
#dataframe_from_series
name = ['macbook air m4', 'nitro v15', 'victus']
brand = ['apple', 'acer', 'hp']
price = [50000, 30000, 25000]

Name = pd.Series(name)
Brand = pd.Series(brand)
Price = pd.Series(price)

#series adapt to dic
frame = {'Name' : Name, 'Brand' : Brand, 'Price' : Price}
df_ex = pd.DataFrame(frame)
df_ex   
```

```python
#export csv
df_ex.to_csv("product.csv",header=True, index=False) #save in the same folder
df_ex.to_csv("product2.csv",header=True, index=True)
df_ex.to_csv("product3.csv",header=False, index=False)
df_ex.to_csv("product4.csv",header=False, index=True)
```

```python
#วันที่แดดร้อน
df_weather[df_weather.Event == "แดดร้อน"]
```

```python
#วันที่ฝนตกและอุณหภูมิมากกว่า 35
df_weather.loc[(df_weather.Event == "ฝนตก") & (df_weather.Temperature>=35)]
```

```python
#วันที่มีเมฆมากหรืออุณหภูมิต่ำกว่า 18
df_weather.loc[(df_weather.Event == "เมฆมาก") | (df_weather.Temperature <= 18)]
```

```python
#isin คำสั่งย่อของ or
df_weather.loc[(df_weather.Temperature.isin([30,25])) & (df_weather.Event=="แดดร้อน")]
```
